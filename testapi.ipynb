*TEST API OUTPUT LEVEL

```python
from ultralytics import YOLO
```

```python
model = YOLO("yolov8_run2.pt")
```

```python
im = Image.open("test2.jpg")
```

```python
# run trained model on uploaded image
r = model(im)
```

```python
r_plotted = r[0].plot()
```

```python
type(r_plotted)
```

```python
print(r_plotted)
```

```python
cv2.imshow("result", r_plotted)
cv2.waitKey(0)
```

```python
cv2.imwrite("im_detected.jpg", r_plotted)
```
